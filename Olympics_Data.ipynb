# Olympics_Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('olympics.csv', skiprows=4)
df
```

####  events in which Edwin Flack win a medal

```python
Edwin = df[(df['Athlete']=='FLACK, Edwin')]
Edwin['Event'].value_counts()
```

####  country that has won the most men's gold medals in Athletics over the years and Sortiing the results alphabetically by the player's names.

```python
gold = df[(df.Medal == 'Gold') & (df.Gender == 'Men') & (df.Sport == 'Athletics')]
gold
```

```python
men = gold.sort_values(by = 'Athlete')
men['NOC'].value_counts()[:10]
```

####   gold medals the Indian Men's have won and name the event also

```python
india = df[(df['NOC']=='IND') & (df['Medal']=='Gold') & (df['Gender']=='Men')]
india
```

```python
india['Event'].value_counts()
```

####  five countries that have won the most medals in recent years (from 2000 to 2008)

```python
medals = df[(df['Edition'] >= 2000 )]
medals
```

####  male gold medal winners for the hockey event over the years. Listing the results starting with the most recent. Showing the Olympic City, Edition, Athlete and the country they represent.

```python
mens = df[(df.Medal == 'Gold') & (df.Gender == 'Men')& (df.Event=='hockey')]
mens
```

```python
mens.sort_values(by = 'Edition', ascending=False )[['City','Edition','Athlete','NOC']]
```
